```python
#Write a Pandas program to create and display a DataFrame from the following dictionary data with index labels.

import numpy as np
import pandas as pd

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 
'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

my_df = pd.DataFrame(exam_data, labels)
my_df
```

```python
#1.Print the three first rows using the head() method.

my_df.head(5)
```

```python
#2.Delete rows with Nan values.

my_df = my_df.dropna()
my_df
```

```python
#4. Write a Pandas program to append a new row 'k' to the DataFrame with
# these values (name: "Suresh", score: 15.5, attempts: 1, qualify: "yes").

exam_data2 = {'name': "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
labels2 = ['k']
my_df2 = pd.DataFrame(exam_data2, labels2)
my_df2 = pd.concat([my_df, my_df2])
my_df2
```

```python
#5. Write a Pandas program to delete the 'attempts' column from the DataFrame.

my_df2.drop('attempts', axis = 1, inplace = True)

my_df2
```

```python
#6. Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.


my_df2['sucess'] = my_df2['score'].apply(lambda x:1 if x > 10 else 0)
my_df2
```

```python
#7. After executing the final DataFrame, export it into a CSV file named "my_data".

my_df2.to_csv('my_data.csv')
my_data = pd.read_csv('my_data.csv')
my_data
```
